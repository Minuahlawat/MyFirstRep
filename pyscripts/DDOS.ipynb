```python
def col_countplot(col, data=data):

    fig, ax = plt.subplots(figsize=(8,4))
    sns.set_style('whitegrid')
   
    # countplot of the given column
    ax = sns.countplot(x=col, hue='label', data=data)
    ax.legend(loc="upper right", labels=('normal', 'attack'))
    ax.set_title("train data")
    plt.xticks(rotation=45)
    plt.show()
```

```python
corr_matrix = x.corr().abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))
variables_to_drop = [column for column in upper.columns if any(upper[column] > 0.60)]
print(variables_to_drop)
```

```python
#Converting/ Scaling the dataset "Data" 
from sklearn.preprocessing import MinMaxScaler
scalar = MinMaxScaler()
features_scalled = scalar.fit_transform(data[column])
data[column] = pd.DataFrame(features_scalled,  columns=data[column].columns)
data.head()
```

```python
#Running feature selection on 'data'
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x, y)
from sklearn.feature_selection import RFE
rf_estimator = RFE(rf, n_features_to_select=18)
rf_estimator = rf_estimator.fit(x,y)
columns = x.columns[rf_estimator.support_]
columns
```

```python
#Running feature selection on 'data2'
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x2, y)
from sklearn.feature_selection import RFE
rf_estimator = RFE(rf, n_features_to_select=18)
rf_estimator = rf_estimator.fit(x2,y)
```

```python
#splitting to the train and test data for both dataset "Data", "Data2"
from sklearn.model_selection import train_test_split
x_train,x_validate,y_train,y_validate = train_test_split(x,y, test_size=0.30,random_state=1, stratify=y)
x2_train,x2_validate,y2_train,y2_validate = train_test_split(x2,y, test_size=0.30, random_state=1, stratify=y)
x3_train,x3_validate,y3_train,y3_validate = train_test_split(x3,y, test_size=0.30,random_state=1, stratify=y)
x4_train,x4_validate,y4_train,y4_validate = train_test_split(x4,y, test_size=0.30, random_state=1,stratify=y)
```

```python
#Using Random Forest Classifier Model with all 4 datasets
rf_base = RandomForestClassifier()
rf_base.fit(x_train,y_train)
rf_base2 = RandomForestClassifier()
rf_base2.fit(x2_train,y2_train)
rf_base3 = RandomForestClassifier()
rf_base3.fit(x3_train,y3_train)
rf_base4 = RandomForestClassifier()
rf_base4.fit(x4_train,y4_train)
```

```python
#Predicting model
y_pred = rf_base.predict(x_validate)
y2_pred = rf_base2.predict(x2_validate)
y3_pred = rf_base3.predict(x3_validate)
y4_pred = rf_base4.predict(x4_validate)
```

```python

from sklearn.metrics import accuracy_score
# Accuracy check of model  
acu_RF=accuracy_score(y_validate,y_pred)
acu_RF2=accuracy_score(y2_validate,y2_pred)
acu_RF3=accuracy_score(y3_validate,y3_pred)
acu_RF4=accuracy_score(y4_validate,y4_pred)
print(acu_RF, acu_RF2, acu_RF3, acu_RF4)
```

```python

from sklearn.metrics import classification_report
print(classification_report(y_validate,y_pred))
print(classification_report(y2_validate,y2_pred))
print(classification_report(y3_validate,y3_pred))
print(classification_report(y4_validate,y4_pred))
```

```python
import matplotlib.pyplot as plt
accu_RF=[acu_RF, acu_RF2, acu_RF3, acu_RF4]
Case=["Orignal","Multi-Collinearity"," RFE-Orignal","RFE-Multi"]
plt.ylim(0.7,1)
plt.bar(Case,accu_RF)
plt.title('Case vs Accuracy for RF')
plt.xlabel('Case')
plt.ylabel('Accuracy')
plt.show()
```
